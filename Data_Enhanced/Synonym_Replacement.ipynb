```python
import os
import re
import jieba

questions_list = []
question_word_list = []

path = r'../processed_question'
for filename in os.listdir(path):
    file = re.match(r'(.*).docx', filename)
    questions_list.append(file.group(1))

for question in questions_list:
    seg_list = list(jieba.cut(question))
    for seg_word in seg_list:
        question_word_list.append(seg_word)

file = r'./question_word.txt'
with open(file, 'w', encoding='utf-8') as f:
    for word in question_word_list:
        f.write(word + '\n')
    
```
